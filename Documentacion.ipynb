```python
import pandas as pd
pd.set_option('display.float_format', '{:.2f}'.format)
import os
result_files = ['TC1.txt_StatisticsResults.txt','TC2.txt_StatisticsResults.txt','TC3.txt_StatisticsResults.txt','TC4.txt_StatisticsResults.txt',
                'TC5.txt_StatisticsResults.txt','TC6.txt_StatisticsResults.txt','TC7.txt_StatisticsResults.txt']

def read_statistics_file(file_path):
    with open(f'./P1/{file_path}', 'r') as file:
        lines = file.readlines()

    stats = {}
    for line in lines[1:]:
        key, value = line.strip().split(': ')
        if key != "Elapsed Time":
            stats[key] = value


    return stats
data = []
for file in result_files:
    stats = read_statistics_file(file)
    
    if stats:
        data.append(stats)

df = pd.DataFrame(data)
df.index = result_files
df
```
